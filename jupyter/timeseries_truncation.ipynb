```julia
ENV["GKS_ENCODING"] = "utf-8"
using Plots
using Plots.PlotMeasures: mm
gr()
```

# Data

```julia
gvfn = Dict(
    "MG"=>(μ = fill(0.1281, 6), σ = fill(0.0182, 6)),
    "MSO"=>(μ = fill(0.0098, 6), σ = fill(0.0012, 6)),
    "ACEA"=>(μ = fill(0.3797, 6), σ = fill(0.0022, 6))
)
```

```julia
rnn = Dict(
    "MG"=>(
        μ = [0.5328, 0.3000, 0.1172, 0.0928, 0.0601, 0.0489],
        σ = [0.0270, 0.0536, 0.0128, 0.0142, 0.0087, 0.0054]),
    "MSO"=>(
        μ = [1.0158, 0.7416, 0.6160, 0.3428, 0.7324, 0.1331],
        σ = [0.0466, 0.0209, 0.0320, 0.0204, 0.1207, 0.0081]), 
    "ACEA"=>(
        μ = [0.5512, 0.5354, 0.5491,0.5267, 0.5518, 0.5517],
        σ = [0.0359, 0.0290, 0.0223, 0.0113, 0.0476, 0.0160])
)

truncation = [1, 2, 4, 8, 16, 32]
```

# Plotting
## Mackey-Glass

```julia
key = "MG"

plot_kwargs=[
    :grid=>false,
    :legend=>nothing,
    :tickdir=>:out,
    :tickfontsize=>14,
    :ylabel=>"NMRSE ± σ",
    :xlabel=>"Truncation (p)",
    :lw=>2,
    :title=>"Mackey-Glass"]

plot(truncation, gvfn[key].μ, ribbon=gvfn[key].σ; plot_kwargs...)
plot!(truncation, rnn[key].μ, ribbon=rnn[key].σ, color=:green)
```
